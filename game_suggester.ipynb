```python
def list_convert(obj):
    list1 = []
    for i in obj:
        list1.append(i)
    result = ''.join(list1)
    return [result]
```

```python
def list_convert1(obj):
    if isinstance(obj, str):
        # If the input is a string, remove the enclosing brackets and split the string by semicolons
        result = obj.strip('[]').split(';')
        return result
    else:
        # If the input is already a list, return it as is
        return obj
```

```python
join_steam["short_description"]=join_steam["short_description"].apply(list_convert)
join_steam["developer"]=join_steam["developer"].apply(list_convert)
join_steam["platforms"]=join_steam["platforms"].apply(list_convert1)
join_steam["genres"]=join_steam["genres"].apply(list_convert1)
join_steam["steamspy_tags"]=join_steam["steamspy_tags"].apply(list_convert1)
```

```python
join_steam["short_description"]=join_steam["short_description"].apply(lambda x:[i.replace(" ","")for i in x])
join_steam["steamspy_tags"]=join_steam["steamspy_tags"].apply(lambda x:[i.replace(" ","")for i in x])
join_steam["platforms"]=join_steam["platforms"].apply(lambda x:[i.replace(" ","")for i in x])
join_steam["genres"]=join_steam["genres"].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000,stop_words='english')
```

```python
vector=cv.fit_transform(join_steam['tags']).toarray()
```

```python
vector
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def stem(text):
    list2=[]
    for i in text.split():
        list2.append(ps.stem(i))
    return " ".join(list2)
```

```python
join_steam['tags']=join_steam['tags'].apply(stem)
```

```python
join_steam['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(game):
    game_index=join_steam[join_steam['name']==game].index[0]
    distance=similarity[game_index]
    game_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:10]
    for i in game_list:
                print(join_steam.iloc[i[0]]['name'])
```

```python
recommend('DOOM II')
```
